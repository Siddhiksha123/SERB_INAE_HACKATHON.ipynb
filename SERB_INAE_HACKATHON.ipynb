```python
import pandas as pd
from simpletransformers.seq2seq import Seq2SeqModel,Seq2SeqArgs
```

```python
model_args = Seq2SeqArgs()
model_args.num_train_epochs = 10
model_args.no_save = True
model_args.evaluate_generated_text = True
model_args.evaluate_during_training = True
model_args.evaluate_during_training_verbose = True

model = Seq2SeqModel(
    encoder_decoder_type="bart",
    encoder_decoder_name="facebook/bart-large",
    args=model_args,
    use_cuda= False,
)
```

```python
df = pd.read_csv('train.csv')
df = df.head(50)
df.rename(columns = {'Abstract':'input_text', 'RHS':'target_text'}, inplace = True)
df
```

```python
train_data =[]
for index, row in df.iterrows():
  train_data.append([row['input_text'], row['target_text']])
print((train_data))
```

```python
train_df = pd.DataFrame(
    train_data, columns=["input_text", "target_text"]
)

```

```python
val_df = df.head(20)
val_df
```

```python
eval_data = []
for index, row in val_df.iterrows():
  eval_data.append([row['input_text'],row['target_text']])
print((eval_data))
```

```python
def count_matches(labels, preds):
    print(labels)
    print(preds)
    return sum(
        [
            1 if label == pred else 0
            for label, pred in zip(labels, preds)
        ]
    )
```

```python
# Train the model
model.train_model(train_df, eval_data=val_df)
```

```python
results = model.eval_model(val_df)
```
